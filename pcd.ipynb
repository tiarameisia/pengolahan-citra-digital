```python
import cv2 #Tiara_Meisia_202331049
import matplotlib.pyplot as plt #Tiara_Meisia_202331049
import numpy as np #Tiara_Meisia_202331049
```

```python
img = cv2.imread ('img_gray.jpg') #Tiara_Meisia_202331049
img.shape #Tiara_Meisia_202331049
```

```python
[baris, kolom] = img.shape[:2] #Tiara_Meisia_202331049
```

```python
rgb = cv2.cvtColor (img, cv2.COLOR_BGR2RGB) #Tiara_Meisia_202331049
```

```python
fig, axs = plt.subplots(2,2, figsize=(15, 5))

#Cara 1
axs[0,0].imshow(rgb) #Tiara_Meisia_202331049
axs[0,1].hist(rgb.ravel(), 256,[0,256]) #Tiara_Meisia_202331049
#Cara 2
hist = cv2.calcHist([rgb], [0], None, [256], [0,256]) #Tiara_Meisia_202331049
axs[1,0].imshow(rgb) #Tiara_Meisia_202331049
axs[1,1].plot(hist) #Tiara_Meisia_202331049

#Menampilkan
plt.show() #Tiara_Meisia_202331049
```

```python
beta = 40 #Tiara_Meisia_202331049
citra_cerah = np.zeros((baris, kolom, 3)) #Tiara_Meisia_202331049

for x in range(baris): #Tiara_Meisia_202331049
    for y in range(kolom): #Tiara_Meisia_202331049
        gyx = rgb[x,y] + beta #Tiara_Meisia_202331049
        citra_cerah[x,y] = gyx #Tiara_Meisia_202331049
        
citra_cerah = citra_cerah.astype(np.uint8) #Tiara_Meisia_202331049

fig, axs = plt.subplots(2,2, figsize=(15,5)) #Tiara_Meisia_202331049
axs[0,0].imshow(rgb) #Tiara_Meisia_202331049
axs[0,1].hist(rgb.ravel(), 256,[0,256]) #Tiara_Meisia_202331049
axs[1,0].imshow(citra_cerah) #Tiara_Meisia_202331049
axs[1,1].hist(citra_cerah.ravel(), 256, [0,256]) #Tiara_Meisia_202331049

plt.show() #Tiara_Meisia_202331049
```

```python
alpha = 1.2 #Tiara_Meisia_202331049
citra_kontras = np.zeros((baris, kolom, 3)) #Tiara_Meisia_202331049

for x in range(baris): #Tiara_Meisia_202331049
    for y in range(kolom): #Tiara_Meisia_202331049
        gyx = rgb[x,y] * alpha #Tiara_Meisia_202331049
        citra_kontras[x,y] = gyx #Tiara_Meisia_202331049
        
citra_kontras = citra_kontras.astype(np.uint8) #Tiara_Meisia_202331049

fig, axs = plt.subplots(2,2, figsize=(15,5)) #Tiara_Meisia_202331049
axs[0,0].imshow(rgb) #Tiara_Meisia_202331049
axs[0,1].hist(rgb.ravel(), 256,[0,256]) #Tiara_Meisia_202331049
axs[1,0].imshow(citra_kontras) #Tiara_Meisia_202331049
axs[1,1].hist(citra_kontras.ravel(), 256, [0,256]) #Tiara_Meisia_202331049

plt.show() #Tiara_Meisia_202331049
```

```python
beta = 10 #Tiara_Meisia_202331049
alpha = 1.2 #Tiara_Meisia_202331049
citra_hasil = np.zeros((baris, kolom, 3)) #Tiara_Meisia_202331049

for x in range(baris): #Tiara_Meisia_202331049
    for y in range(kolom): #Tiara_Meisia_202331049
        gyx = alpha * rgb[x,y] + beta #Tiara_Meisia_202331049
        citra_hasil[x,y] = gyx #Tiara_Meisia_202331049
        
citra_hasil = citra_hasil.astype(np.uint8) #Tiara_Meisia_202331049

fig, axs = plt.subplots(2,2, figsize=(15,5)) #Tiara_Meisia_202331049
axs[0,0].imshow(rgb) #Tiara_Meisia_202331049
axs[0,1].hist(rgb.ravel(), 256,[0,256]) #Tiara_Meisia_202331049
axs[1,0].imshow(citra_hasil) #Tiara_Meisia_202331049
axs[1,1].hist(citra_hasil.ravel(), 256, [0,256]) #Tiara_Meisia_202331049

plt.show() #Tiara_Meisia_202331049
```

```python
beta = 10 #Tiara_Meisia_202331049
alpha = 1.2 #Tiara_Meisia_202331049
citra_negatif = np.zeros((baris, kolom, 3)) #Tiara_Meisia_202331049

for x in range(baris): #Tiara_Meisia_202331049
    for y in range(kolom): #Tiara_Meisia_202331049
        fyx = 255 - rgb[x,y] #Tiara_Meisia_202331049
        citra_negatif[x,y] = fyx #Tiara_Meisia_202331049
        
citra_negatif = citra_negatif.astype(np.uint8) #Tiara_Meisia_202331049

fig, axs = plt.subplots(2,2, figsize=(15,5)) #Tiara_Meisia_202331049
axs[0,0].imshow(rgb) #Tiara_Meisia_202331049
axs[0,1].hist(rgb.ravel(), 256,[0,256]) #Tiara_Meisia_202331049
axs[1,0].imshow(citra_negatif) #Tiara_Meisia_202331049
axs[1,1].hist(citra_negatif.ravel(), 256, [0,256]) #Tiara_Meisia_202331049

plt.show() #Tiara_Meisia_202331049
```
